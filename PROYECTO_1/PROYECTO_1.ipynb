<p style="text-align: justify;">Para la selección de nuestros activos implementamos 3 filtros, el primero es que tengan un rendimiento promedio superior al 6% anual, el segundo parámetro es que tengo cierta regularidad dentro del mercado, que sean activos "comerciables", que fluctuen dentro de la compra y venta de los participantes, esto se observa por medio del volumen registrado, se acepta si tienen un volumen anual superior a 1.00e+10. Y el último paso de selección es que los activos no se encuentre correlacionados, es decir, que sus cambios a lo largo del tiempo sean distintos para mitigar el riesgo sistemático.</p>

+ $Rendimiento$

Utilidad generada en un determinado lapso de tiempo.

$Rendimiento$ $simple$

$$R_t = \frac{S_t - S_{t-1}}{S_{t-1}}$$

$Rendimiento$ $logarítmico$

$$r_t = ln(\frac{S_t}{S_{t-1}})$$

donde:

$S_t$ precio del activo en el tiempo $t$

+ $Volumen$

Cantidad de títulos que se negocian en cada sesión o cantidad de dinero que mueve cada día un activo (acciones, bonos, por ejemplo)

+ $Correlación$

<p style="text-align: justify;">La correlación entre dos activos es una medida estadística que nos muestra el grado de relación entre ambos y cuyo valor se encuentra en el intervalo entre 1 y -1. Cuando hablamos de valores o activos financieros, el coeficiente de correlación, representa el grado de relación entre los movimientos de los precios de los diferentes activos incluidos en la cartera. Si su valor es 1, significa que los precios se mueven a la par (una modificación en el precio de un activo implica otra del mismo sentido en el precio del otro), una correlación de -1 significa que los precios se mueven en direcciones opuestas. Si su valor es cero significa que los precios se encuentran linealmente incorrelacionados, es decir, el movimiento del precio de un activo no tiene efecto sobre el movimiento del precio de otro activo.</p>

$$p = \frac{σ_{xy}}{σ_xσ_y}$$

donde:

$σ_{xy}$ es la covarianza entre los activos

$σ_x$ desviación estandar del activo x

$σ_y$ desviación estandar del activo y


+ $Simulación$ $Montecarlo$

<p style="text-align: justify;">Es un método numérico que permite resolver problemas físicos y matemáticos mediante la simulación de variables aleatorias. Lo vamos a considerar aquí desde un punto de vista didáctico para simular los pesos por activo, los rendimientos y la desviacion estándar futura.</p>

+ $Sharpe$

<p style="text-align: justify;">El Ratio de Sharpe mide el rendimiento en exceso (o prima de riesgo) por unidad de desviación típica en los activos de inversión. Así, el Ratio de Sharpe representa como la rentabilidad de un activo, cartera o portafolio compensa el riesgo que se asume al invertir en el. Al comparar dos activos frente a un punto de referencia común (índice del mercado), el que tiene un mayor Ratio de Sharpe proporciona una mejor rentabilidad para el mismo riesgo.</p>

\begin{equation*}
S_p =\frac{E_p - R_f}{σ_p} 
\end{equation*}


+ $Black$ & $Scholes$

<p style="text-align: justify;">En 1973 Fisher Black, Myron Scholes y Robert Merton lograron uno de los mayores avances en la valuación de opciones hasta ese momento, conocido como el modelo de Black-Scholes, que ha tenido una gran influencia en la manera en que los agentes valúan y cubren opciones. Ha sido también un punto de referencia para el desarrollo y éxito de la ingeniería financiera desde entonces. </p>

La fórmula Black-Scholes para calcular el precio teórico de una opción de compra (call) y de venta (put) europeas son:

$$C = S_tN(d1)-N(d2)Ke^(rt)$$
$$P = N(-d2)Ke^(rt)-S_tN(-d1)$$

en donde:
$C$ es el valor teórico de la opción call,
$P$ es el valor teórico de la opción put,
$S_t$ el precio de la acción en la que la opción está denominada,
$K$ el precio del ejercicio de la opción,
$t$ el número de años que le restan de vida a la opción,
$e$ la base de los logaritmos naturales,
$r$ la tasa libre de riesgo,  
$σ$ la volatilidad anual (desviación estándar) del rendimiento de la acción, 
$N(d1)$ la probabilidad de que el precio se ejerza,
$N(d2)$ la probabilidad

donde:
$$ d1 = \frac{ln(S_t/K)+(r+σ^2/2)T}{σ\sqrt{T}}$$

$$ d2 = \frac{ln(S_t/K)+(r-σ^2/2)T}{σ\sqrt{T}} = d1-σ\sqrt{T} $$

<p style="text-align: justify;">El Ratio de Sharpe (también conocido como el índice de Sharpe) mide el rendimiento en exceso (o prima de riesgo) por unidad de desviación típica en los activos de inversión.

Así, el Ratio de Sharpe representa como la rentabilidad de un activo, cartera o portafolio compensa el riesgo que se asume al invertir en el. Al comparar dos activos frente a un punto de referencia común (índice del mercado), el que tiene un mayor Ratio de Sharpe proporciona una mejor rentabilidad para el mismo riesgo.</p>

**Fórmula de Sharpe Ratio**

$$S_{p} = \frac{E_{p}-R_{f}}{σ_{p}}$$.

Donde:


$S_{p} = Sharpe Ratio$


$E_{p} = Rendimiento del portafolio$


$R_{f} = Tasa libre de riesgo$


$σ_{p} = Desviación estandar del portafolio$

<p style="text-align: justify;">La fórmula del Ratio de Sharpe supone por tanto una medida del rendimiento de una inversión, calculada dividiendo el exceso de rentabilidad (que está por encima del rendimiento de una inversión libre de riesgo como por ejemplo los bonos del Tesoro) por la cantidad de riesgo adicional que se asume medido como la desviación típica de los rendimientos.</p>
<p style="text-align: justify;">Un portafolio con un Sharpe Ratio mayor a 1 es considerado como un buen portafolio para invertir.</p>

# 3. Coberturas

<p style="text-align: justify;">Las coberturas son contratos de opción que dan al propietario derecho, pero no la obligación de vender una cantidad de un activo a un precio y plazo determinados.</p>

<p style="text-align: justify;">Las coberturas protegen las finanzas públicas en caso de una caída abrupta de los precios (precios del crudo en este caso)</p>

```python
data=portfolio_bond3.get_historical_closes(['BNO'], '2016-10-10', '2017-10-10')
data.plot(figsize=(8,6));
```

```python
daily_returns=portfolio_bond3.calc_daily_returns(data)
daily_returns.plot(figsize=(8,6));
```

```python
huber = sm.robust.scale.Huber()
returns_av, scale = huber(daily_returns)
mu=daily_returns.mean().BNO
sigma=daily_returns.std().BNO
ndays = 360
ntraj=10000
K1=13
K2=16
```

```python
mu=daily_returns.mean().BNO
sigma=daily_returns.std().BNO
dates=pd.date_range('20171009',periods=ndays)
simret = pd.DataFrame(sigma*np.random.randn(ndays,ntraj)+mu,index=dates)
simdata=(data.loc['2017-10-11',:].BNO)*np.exp(simret.cumsum())
ST=simdata.tail(1)
ind=np.digitize(ST.values[0],np.array([0, K1, K2, ST.values.max()+1]))
freq=(np.array([sum(ind==1),sum(ind==2),sum(ind==3)])/ntraj)*100
```

```python
bar_labels = ['ST<K1', 'K1<ST<K2', 'ST>K2']
x_pos = list(range(len(bar_labels)))
plt.bar(x_pos,freq);
plt.grid();
plt.xticks(x_pos, bar_labels);
```

```python
values,indices=np.histogram(daily_returns,bins=250)
values=values.astype(np.float32)
weights=values/np.sum(values)
ret=np.random.choice(indices[1:],ndays*ntraj,p=weights)
#
dates=pd.date_range('20171011',periods=ndays)
simret = pd.DataFrame(ret.reshape((ndays,ntraj)),index=dates)
simdata=(data.loc['2017-10-09',:].BNO)*np.exp(simret.cumsum())
ST=simdata.tail(1)
ind=np.digitize(ST.values[0],np.array([0, K1, K2, ST.values.max()+1]))
freq=(np.array([sum(ind==1),sum(ind==2),sum(ind==3)])/ntraj)*100
```
